* ¿Cómo la convertiríamos a escala de grises? (sin usar funciones... ¡operando con el array!)

```python
# SOL1: RECORREMOS LOS PIXELES DE LA IMAGEN. ESTA SOLUCIÓN NO ESTA RECOMENDADA
# POR QUE NO ES OPTIMA YA QUE np.array NOS PERMITE OPERAR SIN BUCLES
#PARECE SER QUE EL GREY SACLE SE CONSIGUE CALCULANDO EL VALOR MEDIO
#DE LA SUMA DE LOS VALORES RGB QUE COMPONEN UN PIXEL Y SUSTITUENDO EL
#EL VALOR DE LOS 3 POR LA MEDIA CALCULADA.

def grayscale(imagen):
    imagen_grises=PIL.Image.new(imagen.mode, imagen.size)
    width, height = imagen.size
    for i in range(0, width):
            for j in range(0, height):
                pixel=imagen.getpixel((i,j))
                red= pixel[0]
                green= pixel[1]
                blue= pixel[2]
                gray = int((red + green + blue)/3)
                imagen_grises.putpixel((i,j),(gray,gray,gray))
    imagen_grises.show()
                            
grayscale(imagen)

#SOL2: ESTA SOLUCIÓN NO ES VÁLIDA YA QUE SE NOS PIDE OPERAR SOBRE EL ARRAY
#DE LA IMAGEN
#imagen.convert('L') 

```

### Aplicando filtros a imagenes

Los algoritmos de computer vision realizan un filtrado de las imágenes para quedarse con sus principales características y para crear de una misma imagen un conjunto más amplio con el que entrenar al modelo. En este ejercicio se pide aplicar el filtrado de la media (para difuminar una imagen) y la mediana (para disminuir el ruido). Más información acerca de los diferente tipos de filtros [aqui](https://medium.com/image-vision/noise-filtering-in-digital-image-processing-d12b5266847c).

* Primero vamos aplicar el filtro de la media a la imagen original de Pikachu. Para ello tenemos que ir recorriendo cada pixel de la imagen y generando una ventana (kernel) de tamaño n (el tamaño de la ventana va a ser simétrico en las dos dimensiones) que este centrada en el pixel, calcule la media de los elementos en esa ventana y reemplace el valor original del pixel por la media. En este ejemplo tendriamos una ventana de 3x3:
<br/>
<img src="mean_filter.png" style="width: 600px">

* MIO se utiliza esto para coputer vision (deep learning). Esto lo hace TensorFlow, que ya lo hace automaticamente pero es interesante que lo entendamos como lo hace. La ventana es el grado de difuminación

```python
#Función que tiene como entrada el tamaño de la ventana:
#Rellenar la función para que aplique el filtro de la media sobre la imagen de Pikachu
#n es el tamaño de la ventana. A la función la pasamos el nombre o url de la imagen
def mean_filter(n,imagen_name):
    img = imageio.imread(imagen_name)

    R = img[:,:,0]
    G = img[:,:,1]
    B = img[:,:,2]

    r_d = np.zeros(shape=(img.shape[0],img.shape[1]))
    g_d = np.zeros(shape=(img.shape[0],img.shape[1]))
    b_d = np.zeros(shape=(img.shape[0],img.shape[1]))
#Tenemos que ir recorriendo cada pixel de la imagen. Para ellos usamos dos loops:
    for i in range(img.shape[0]):
        for j in range(img.shape[1]):
            
            x_min=(i-n)    
            x_max=(i+n)    
            y_min=(j-n)    
            y_max=(j+n)  
            
            if x_min<0:
                x_min=0
            if y_min<0:
                y_min=0 

            if x_max>img.shape[0]:
                x_max=img.shape[0]

            if y_max>img.shape[1]:
                y_max=img.shape[1] 
                
            r_d[i,j]=R[x_min:x_max+1,y_min:y_max+1].mean()
            g_d[i,j]=G[x_min:x_max+1,y_min:y_max+1].mean()
            b_d[i,j]=B[x_min:x_max+1,y_min:y_max+1].mean()
            
    img_mean=np.dstack([r_d,g_d,b_d]) 
    
    img_mean=img_mean.astype(np.uint8)  
    
    return img_mean
```

```python
img_mean_filter_20=mean_filter(5,'https://bit.ly/33iZhPz')
PIL.Image.fromarray(img_mean_filter_20)
```

* El filtro de la mediana lo vamos a aplicar sobre una imagen en escala de grises que tiene simulado una gran cantidad de ruido. Este es el ejemplo de como actuaría el filtro de la mediana sobre un pixel y con una ventana de tamaño 3.
<br/>
<img src="median_filter.png" style="width: 600px">

* MIO Con esto quitamos el ruido de la imagen.

```python
#Función que tiene como entrada el tamaño de la ventana y aplica el filtro de la mediana:
#Rellenar la función para que aplique el filtro de la mediana sobre la imagen noise.png
#n es el tamaño de la ventana. A la función la pasamos el nombre o url de la imagen
def median_filter(n,imagen_name):
    img = imageio.imread(imagen_name)
    img_w_o_noise = np.zeros(shape=(img.shape[0],img.shape[1]))
    for i in range(img.shape[0]): # RECORRIDO FILAS
        for j in range(img.shape[1]): # RECORRIDO COLUMNAS
            #AQUI FIJAMOSEL (XX,YY) QUE COMPRENA EL MARCO A TRATAR TENIENDO EN CUENTA EL TAMAÑO DE LA VENTANA
            x_min=(i-n)    
            x_max=(i+n)    
            y_min=(j-n)    
            y_max=(j+n)  

            if x_min<0:
                x_min=0
            if y_min<0:
                y_min=0 

            if x_max>img.shape[0]:
                x_max=img.shape[0]

            if y_max>img.shape[1]:
                y_max=img.shape[1] 

            img_w_o_noise[i,j]=np.median(img[x_min:x_max+1,y_min:y_max+1])

    img_w_o_noise=img_w_o_noise.astype(np.uint8)        
    return img_w_o_noise
```
